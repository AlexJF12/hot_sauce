# Word Vectorized on Description, Ingredients

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
df = pd.read_csv('./hot_sauce_clean.csv')
df = df[['names', 'mfc', 'descrip', 'ing', 'heat_lvl']]
df['ing'] = df['ing'].apply(lambda x: x.replace('Ingredients:',''))
```

```python
df.tail()
```

### Count Vectorizer on Description

```python
from sklearn.feature_extraction.text import CountVectorizer

cvec = CountVectorizer(stop_words='english', lowercase=True)
cvec.fit(df['descrip'])
```

```python
df_cvec = pd.DataFrame(cvec.transform(df['descrip']).todense(),
                   columns=cvec.get_feature_names())
```

```python
df_cvec.shape
```

```python
# print top occuring words in descriptions
df_cvec.mean().sort_values(ascending = False)[:10]
```

### Term-Frequency Inverse Documenct Frequency on Description

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tvec = TfidfVectorizer(stop_words='english', lowercase=True, ngram_range=(1,1))

tvec.fit(df['descrip'])
df_tvec  = pd.DataFrame(tvec.transform(df['descrip']).todense(),
                        columns=tvec.get_feature_names())
```

```python
df_tvec.shape
```

```python
# list of top 50 most 'important' words
words_to_keep = df_tvec.mean().sort_values(ascending = False)[:50].index
```

```python
words_to_keep
```

```python
# combine heat_lvl dummies and df_tvec words
df_wide = pd.concat([pd.get_dummies(df['heat_lvl']), df_tvec], axis = 1)
```

### Model on Word count / tf-idf values

```python
# kmeans model it
from sklearn.cluster import KMeans

km = KMeans(n_clusters=5)
km.fit(df_tvec)

# make groups
groups = km.predict(df_tvec)
```

```python
# pick some words...

plt.scatter(df_tvec['garlic'], df_tvec['tomato'], c=groups)
plt.show()
```

### PCA to reduce feature count

```python
from sklearn.decomposition import PCA
pca = PCA().fit(df_tvec)
```

```python
# pca.explained_variance_ratio_
```

```python
# plot the explained variance of each level of PCA

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(range(1,20), pca.explained_variance_ratio_[:19], lw=2)
ax.scatter(range(1, 20), pca.explained_variance_ratio_[:19], s=100, c = 'b')
ax.set_title('explained variance of components')
ax.set_xlabel('principal component')
ax.set_ylabel('explained variance');
```

```python
# plot cumulative variance explained from PCA lavels

cum_var_exp = np.cumsum(pca.explained_variance_ratio_[:19])*100
plt.figure(figsize=(9,7))
component_number = range(1,20)
plt.plot(component_number, cum_var_exp, lw=7)
plt.axhline(y=0, linewidth=5, color='grey', ls='dashed')
plt.axhline(y=100, linewidth=3, color='grey', ls='dashed')
ax = plt.gca()
ax.set_xlim([1,20])
ax.set_ylim([-5,105])
ax.set_ylabel('cumulative variance explained', fontsize=16)
ax.set_xlabel('component', fontsize=16)  
ax.set_title('component vs cumulative variance explained\n', fontsize=20);
```

```python
# transform data to PCA vectors
pca_df = pd.DataFrame(pca.transform(df_tvec))
pca_df.head()
```

```python
# plot the first and second PCA components
# colors are kmeans model unsupervised 'group'

fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(pca_df.iloc[:,0], pca_df.iloc[:,1], s=100, c=groups)
ax.set_xlabel('principal component 1')
ax.set_ylabel('principal component 2');
```

### Ingredients by individual item

```python
# split ingredients column by individual ingredient

# split into lists
ing_split = []
for i in df['ing']:
    ing_split.append(i.split(','))

# clean items in each list
ing_split2 = []
for i in ing_split:
    lst = []
    for k in i:
        lst.append(k.strip()
                   .lower()
                   .replace('.',''))
    ing_split2.append(lst)
```

```python
# create dummies if ingredient is in product
ing_dum = pd.get_dummies(pd.Series(ing_split2).apply(pd.Series).stack()).sum(level=0)
```

```python
# see top ingredients by split method
ing_dum.mean().sort_values(ascending = False)[:10]
```

## KMeans model

```python
km = KMeans(n_clusters=5)
km.fit(ing_dum)

# make groups
groups = km.predict(ing_dum)
groups
```
